```python
import pandas as pd

# Load dataset
steps_table = pd.read_csv("hourlySteps_merged.csv")

# Convert to datetime
steps_table['ActivityHour'] = pd.to_datetime(steps_table['ActivityHour'])

# Create DateTime table
datetime_table = pd.DataFrame()

datetime_table['DateTime'] = steps_table['ActivityHour']

# Date attributes
datetime_table['Date'] = datetime_table['DateTime'].dt.date
datetime_table['Year'] = datetime_table['DateTime'].dt.year
datetime_table['DayName'] = datetime_table['DateTime'].dt.day_name()

# Combine MonthName and Day
datetime_table['Month_Day'] = (
    datetime_table['DateTime'].dt.month_name() + " " +
    datetime_table['DateTime'].dt.day.astype(str)
)

# Time attributes
datetime_table['Hour'] = datetime_table['DateTime'].dt.hour
datetime_table['Minute'] = datetime_table['DateTime'].dt.minute
datetime_table['Second'] = datetime_table['DateTime'].dt.second

# Remove duplicates
datetime_table = datetime_table.drop_duplicates()

# Save table
datetime_table.to_csv("datetime_table.csv", index=False)

# Display output
datetime_table.head(5)
```

```python
import pandas as pd

# Load dataset
weight_log_information_table = pd.read_csv("weightLogInfo_merged.csv")

# Remove unnecessary columns safely
weight_log_information_table = weight_log_information_table.drop(columns=['LogId','IsDay'], errors='ignore')

# Keep only 2 decimal values
weight_log_information_table = weight_log_information_table.round(2)

# Change Id format
weight_log_information_table['Id'] = [
    'user_' + str(i).zfill(4)
    for i in range(1, len(weight_log_information_table)+1)
]

# Create BMI groups
weight_log_information_table['WeightGroup'] = pd.cut(
    weight_log_information_table['BMI'],
    bins=[0,18.5,25,30,100],
    labels=['Underweight','Normal Weight','Overweight','Obese']
)

# Save table
weight_log_information_table.to_csv("weight_log_information_table.csv", index=False)

# Show output
weight_log_information_table.head(5)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
activity_table = pd.read_csv("dailyActivity_merged.csv")

# Convert date
activity_table['ActivityDate'] = pd.to_datetime(activity_table['ActivityDate'])

# Plot
plt.figure()
plt.bar(activity_table['ActivityDate'], activity_table['TotalSteps'])

plt.xlabel("Date")
plt.ylabel("Total Steps")
plt.title("Daily Total Steps")

plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

activity_table = pd.read_csv("dailyActivity_merged.csv")

plt.figure()
plt.scatter(activity_table['TotalSteps'], activity_table['Calories'])

plt.xlabel("Total Steps")
plt.ylabel("Calories Burned")
plt.title("Steps vs Calories")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

activity_table = pd.read_csv("dailyActivity_merged.csv")

activity_minutes = [
    activity_table['VeryActiveMinutes'].sum(),
    activity_table['SedentaryMinutes'].sum()
]

labels = ["Very Active Minutes", "Sedentary Minutes"]

plt.figure()
plt.pie(activity_minutes, labels=labels, autopct='%1.1f%%')

plt.title("Activity Level Distribution")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
activity = pd.read_csv("dailyActivity_merged.csv")

# Create scatter plot
sns.scatterplot(x=activity["TotalSteps"], y=activity["Calories"])

plt.title("Steps vs Calories Burned")
plt.xlabel("Total Steps")
plt.ylabel("Calories Burned")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
activity = pd.read_csv("dailyActivity_merged.csv")

# Calculate average minutes
active_minutes = activity["VeryActiveMinutes"].mean()
sedentary_minutes = activity["SedentaryMinutes"].mean()

# Create bar chart
labels = ["Very Active Minutes", "Sedentary Minutes"]
values = [active_minutes, sedentary_minutes]

plt.bar(labels, values)

plt.title("Active vs Sedentary Behavior")
plt.ylabel("Average Minutes")
plt.xlabel("Activity Type")

plt.show()
```
